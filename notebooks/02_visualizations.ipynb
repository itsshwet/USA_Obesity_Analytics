```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cleaned_obesity_data.csv")
df.head()
```

```python
# Obesity Trend Over Years
```

```python
trend = df.groupby("YearStart")["Data_Value"].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(trend["YearStart"], trend["Data_Value"], marker="o")
plt.title("Obesity Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Average Obesity (%)")
plt.grid(True)
plt.show()
```

```python
# Top 10 States with Highest Obesity
```

```python
state_rank = df.groupby("LocationDesc")["Data_Value"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,5))
bars = state_rank.plot(kind="bar", color="slategray")

plt.title("Top 10 States by Obesity (%)")
plt.xlabel("State")
plt.ylabel("Average Obesity (%)")
plt.xticks(rotation=45)

# add labels
for i, v in enumerate(state_rank.values):
    plt.text(i, v + 0.2, f"{v:.2f}", ha='center', fontsize=9)

plt.show()
```

```python
# Obesity by Gender
```

```python
gender = df.groupby("Sex")["Data_Value"].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
bars = gender.plot(kind="bar", color="lightcoral")

plt.title("Obesity by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Obesity (%)")
plt.xticks(rotation=45)

for i, v in enumerate(gender.values):
    plt.text(i, v + 0.2, f"{v:.2f}", ha='center')

plt.show()
```

```python
# Obesity by Age Group
```

```python
age_grp = df.groupby("Age")["Data_Value"].mean().sort_values(ascending=False)

plt.figure(figsize=(12,5))
bars = age_grp.plot(kind="bar", color="lightgreen")

plt.title("Obesity by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Obesity (%)")
plt.xticks(rotation=45)

for i, v in enumerate(age_grp.values):
    plt.text(i, v + 0.2, f"{v:.2f}", ha='center')

plt.show()
```

```python
# Obesity by Education Level
```

```python
edu = df.groupby("Education")["Data_Value"].mean().sort_values(ascending=True)

plt.figure(figsize=(12,5))
bars = edu.plot(kind="barh", color="mediumpurple")

plt.title("Obesity vs Education Level")
plt.xlabel("Average Obesity (%)")
plt.ylabel("Education Level")

for i, v in enumerate(edu.values):
    plt.text(v + 0.2, i, f"{v:.2f}", va='center')

plt.show()
```

```python
# Obesity by Income Category
```

```python
income = df.groupby("Income")["Data_Value"].mean().sort_values(ascending=True)

plt.figure(figsize=(12,5))
bars = income.plot(kind="barh", color="goldenrod")

plt.title("Obesity vs Income Level")
plt.xlabel("Average Obesity (%)")
plt.ylabel("Income Bracket")

for i, v in enumerate(income.values):
    plt.text(v + 0.2, i, f"{v:.2f}", va='center')

plt.show()
```
